```python
import os
import json
import boto3
import pandas as pd
#from io import StringIO
pd.set_option("display.max_columns", 50)
```

```python
aws_acces_key_id = os.getenv("AWS_ACCESS_KEY")
aws_secret_key = os.getenv("AWS_SECRET_KEY")
```

```python
def read_json_from_s3(bucket: str, path: str) -> json:
    """Downloads a json file from an S3 bucket
    
    Parameters:
    -----------
    bucket : str
        The bucket where the files at.
    path : str
        The folders to the files.
        
    Returns:
    --------
    json
        A downloaded json file.
    """
    
    s3 = boto3.client("s3", aws_access_key_id=aws_acces_key_id, aws_secret_access_key=aws_secret_key)

    for file in  s3.list_objects(Bucket=bucket, Prefix=path)['Contents']:
        key = file["Key"]

        if key.split("/")[-1].strip() != "":
            if key.split(".")[1] == "json":
                
                filename = key.split("/")[-1]

                response = s3.get_object(Bucket=bucket, Key=key)
                content = response["Body"]
                data_json = json.loads(content.read())

    return data_json
```

```python
s3 = boto3.client("s3", aws_access_key_id=aws_acces_key_id, aws_secret_access_key=aws_secret_key)
bucket = "cubix-chicago-taxi-goga"
path = "raw_data/processed/taxi_data/"    

taxi_trips_data_json = read_json_from_s3(bucket=bucket, path=path)
taxi_trips_data_json
```
